```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df_op = pd.read_csv('/Users/jayboyd/Desktop/CTBA/M2/OpidsVA.csv')

fig, ax = plt.subplots()
ax.scatter(df_op['MedianHouseholdIncome'], df_op['FPOO-Rate'], alpha=0.5, c = 'b')  
fig.suptitle('Opiods Rate Vs. Median Household Income')   # Graph title
ax.xaxis.set_label_text('Median Household Income')  # x-axis caption
ax.yaxis.set_label_text('Opioid Use Rate')       # y-axis caption
fig.set_size_inches(7,5)
plt.show()
```

It appears that opioid use rate is heavily concentrated in the lower median household income bracket based on the graph above.

```python
df_op = pd.read_csv('/Users/jayboyd/Desktop/CTBA/M2/OpidsVA.csv')

fig, ax = plt.subplots()
ax.scatter(df_op['FPOO-Rate'], df_op['FFHO-Rate'], alpha=0.5, c = 'b')  
fig.suptitle('Opiods Rate Vs. Heroin Rate')   # Graph title
ax.xaxis.set_label_text('Opioid Use Rate')  # x-axis caption
ax.yaxis.set_label_text('Heroin Use Rate')       # y-axis caption
fig.set_size_inches(7,5)
plt.show()
```

There does seem to be a slight linear relation between opioid use, though there seems to be a significant number of people who exclusively overdoes on one or the other. As a user overdoses more on opioids, they tend to be quite likely to overdose on heroin as well.

```python
df_cars = pd.read_csv('/Users/jayboyd/Desktop/CTBA/M2/TOTALNSA.csv')

fig, ax = plt.subplots()
ax.scatter(df_cars['DATE'], df_cars['TOTALNSA'], alpha=0.5, c = 'b')  
fig.suptitle('Car Sales In USA Over Time')
ax.xaxis.set_label_text('Date')
ax.yaxis.set_label_text('Total NSA')
fig.set_size_inches(7,5)
plt.show()
```
